1. 請嘗試使用 scikit-learn 中「不同基本模型」，並且進行比較結果？
此處使用隨機森林的方式

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
df_train = df.copy()

#填補空值 以baseline的補法

df_train['Age'].fillna(df_train['Age'].median(), inplace=True)
  #cabin	Cabin number 共缺失687 母體是891 缺失值太多 不採用
  #embarked	Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton | 用轉碼的方式替代字母 並用轉碼的眾數填補缺失值
df_train['Embarked'] = df_train['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
df_train['Embarked'].fillna(df_train['Embarked'].mode()[0], inplace=True)

  #性別要轉碼   female: 0, male: 1
df_train['Sex'] = df_train['Sex'].map({'female': 0, 'male': 1})
```

```python
#將不需要的欄位drop off
df_train.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df_train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

columns_X = set(df_train.columns) - {'Survived'}
columns_y = ['Survived']
```

```python
print(columns_X)
columns_X = df_train[list(columns_X)]
columns_y = df_train[columns_y]
```

```python
X_train, X_test, y_train, y_test = train_test_split(columns_X, columns_y, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
# 評估模型
accuracy = accuracy_score(y_test, y_pred)
print("準確率:", accuracy)
print(classification_report(y_test, y_pred))

#相較於baseline 有更好的準確率。  0.8268>擇一[0.78212291 0.78651685 0.78089888 0.76966292 0.81460674]
```

```python
#XGBoost 適合處理特徵工程豐富且需要強大預測能力的問題
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
df_train_XG = df.copy()
#填補空值 以baseline的補法

df_train_XG['Age'].fillna(df_train_XG['Age'].median(), inplace=True)
  #cabin	Cabin number 共缺失687 母體是891 缺失值太多 不採用
  #embarked	Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton | 用轉碼的方式替代字母 並用轉碼的眾數填補缺失值
df_train_XG['Embarked'] = df_train_XG['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
df_train_XG['Embarked'].fillna(df_train_XG['Embarked'].mode()[0], inplace=True)
  #性別要轉碼   female: 0, male: 1
df_train_XG['Sex'] = df_train_XG['Sex'].map({'female': 0, 'male': 1})

#將不需要的欄位drop off
df_train_XG.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

columns_X = set(df_train_XG.columns) - {'Survived'}
columns_y = ['Survived']

columns_X_XG = df_train_XG[list(columns_X)]
columns_y_XG = df_train_XG[columns_y]

X_train, X_test, y_train, y_test = train_test_split(columns_X_XG, columns_y_XG, test_size=0.3, random_state=42)

# 使用 XGBoost 建立模型
model = XGBClassifier()
model.fit(X_train, y_train)

# 預測與評估
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("XGBoost 測試集的準確率: {:.2f}%".format(accuracy * 100))
#XGBoost 測試集的準確率: 78.73%
```

```python
#PyTorch 適合需要深度學習的問題，尤其是當數據量大且非線性關係較複雜時
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
df_train_Py = df.copy()
#填補空值 以baseline的補法

df_train_Py['Age'].fillna(df_train_Py['Age'].median(), inplace=True)
  #cabin	Cabin number 共缺失687 母體是891 缺失值太多 不採用
  #embarked	Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton | 用轉碼的方式替代字母 並用轉碼的眾數填補缺失值
df_train_Py['Embarked'] = df_train_Py['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
df_train_Py['Embarked'].fillna(df_train_Py['Embarked'].mode()[0], inplace=True)
  #性別要轉碼   female: 0, male: 1
df_train_Py['Sex'] = df_train_Py['Sex'].map({'female': 0, 'male': 1})

#將不需要的欄位drop off
df_train_Py.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

columns_X = set(df_train_Py.columns) - {'Survived'}
columns_y = ['Survived']

columns_X_Py = df_train_Py[list(columns_X)]
columns_y_Py = df_train_Py[columns_y]

X_train, X_test, y_train, y_test = train_test_split(columns_X_Py, columns_y_Py, test_size=0.3, random_state=42)

# 確保所有資料都是 NumPy 陣列格式
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long).view(-1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.long).view(-1)



# 建立簡單的神經網路模型
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(X_train_tensor.shape[1], 32)
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16, 2)  # 二分類輸出

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# 初始化模型、損失函數和優化器
model = SimpleNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 訓練模型
num_epochs = 50
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(X_train_tensor)  # 確保 outputs 的形狀是 [batch_size, num_classes]

    # 計算損失
    loss = criterion(outputs, y_train_tensor)

    # 反向傳播與優化
    loss.backward()
    optimizer.step()

    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')


with torch.no_grad():
    outputs = model(X_test_tensor)
    _, predicted = torch.max(outputs, 1)
    accuracy = accuracy_score(y_test_tensor, predicted)
    print("PyTorch 測試集的準確率: {:.2f}%".format(accuracy * 100))
#PyTorch 測試集的準確率: 62.69%
```

```python
#TensorFlow適合需要深度學習的問題，尤其是當數據量大且非線性關係較複雜時
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 載入與前處理資料（與前面範例相同）
df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
df_train_ten = df.copy()
#填補空值 以baseline的補法

df_train_ten['Age'].fillna(df_train_ten['Age'].median(), inplace=True)
  #cabin	Cabin number 共缺失687 母體是891 缺失值太多 不採用
  #embarked	Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton | 用轉碼的方式替代字母 並用轉碼的眾數填補缺失值
df_train_ten['Embarked'] = df_train_ten['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
df_train_ten['Embarked'].fillna(df_train_ten['Embarked'].mode()[0], inplace=True)
  #性別要轉碼   female: 0, male: 1
df_train_ten['Sex'] = df_train_ten['Sex'].map({'female': 0, 'male': 1})

#將不需要的欄位drop off
df_train_ten.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

columns_X = set(df_train_ten.columns) - {'Survived'}
columns_y = ['Survived']

columns_X_ten = df_train_ten[list(columns_X)]
columns_y_ten = df_train_ten[columns_y]

X_train, X_test, y_train, y_test = train_test_split(columns_X_ten, columns_y_ten, test_size=0.3, random_state=42)

# 建立 TensorFlow 模型
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(2, activation='softmax')  # 二分類輸出
])

# 編譯模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 訓練模型
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)

# 評估模型
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("TensorFlow 測試集的準確率: {:.2f}%".format(accuracy * 100))
#TensorFlow 測試集的準確率: 74.63%
```
